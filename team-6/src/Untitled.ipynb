```python
import sys
sys.path.insert(0, "/path/to/cardioagent_demo")

from planner import CardioAgentPlanner
from IPython.display import display, Image, Markdown, HTML

planner = CardioAgentPlanner(
    qwen_api_url="http://localhost:8000/v1",
    lingshu_api_url="http://localhost:8001/v1",
)

state = planner.run(
    patient_id="P001",
    ecg_path="/data/ecg/00001_hr",
    clinical_notes="55yo male, chest pain",
)

# show the process of thinking
for step in state.thinking_steps:
    icon = {"success": "✅", "error": "❌", "skipped": "⏭️"}.get(step.status, "⏳")
    color = {"success": "green", "error": "red", "skipped": "gray"}.get(step.status, "orange")
    display(HTML(
        f'<div style="border-left:3px solid {color}; padding:8px 12px; margin:4px 0; '
        f'background:#1a1a2e; border-radius:0 6px 6px 0;">'
        f'<b>{icon} {step.step_name}</b> ({step.duration_s}s)<br>'
        f'<span style="color:#888">{step.detail}</span></div>'
    ))
    for sub in step.substeps:
        s_icon = "✓" if sub.get("status") == "success" else "✗"
        display(HTML(f'<div style="padding-left:24px; color:#666; font-size:12px">{s_icon} {sub.get("step","")}: {sub.get("detail","")}</div>'))

# show the image
if state.images.get("ecg_waveform"):
    display(Markdown("### 12-Lead ECG"))
    display(Image(filename=state.images["ecg_waveform"], width=800))

if state.images.get("mri_montage"):
    display(Markdown("### Cardiac MRI Montage"))
    display(Image(filename=state.images["mri_montage"], width=800))

# show the report
display(Markdown("### Diagnostic Report"))
display(Markdown(state.final_report))
```
